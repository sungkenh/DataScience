# Uncovering Machine Learning

* 머신 러닝은 데이터를 사용하고, 알고리즘을 생성하는 프로그램을 가르치기 위한 기술임
* 컴퓨터 과학 분야의 인공 지능에서 유래됨
* 통계학과 수학적인 최적화와 밀접한 관련이 있음
* 머신 러닝은 e-mail 스팸 필터, 검색엔진, 언어 번역, 컴퓨터 비전 등의 응용 분야가 있음
* 이 챕터에서 사용되는 몇 가지 용어의 정리
    * __Features__ : 결과를 정의하는데 도움을 주는 구분적 특성
    * __Samples__ : 처리하기 위한 아이템을 말함, ex)문서, 이미지, 오디오, csv파일 등
    * __Feature vector__ : 수치적 벡터를 말하며 n-차원의 벡터나 몇몇 오브젝트를 말함
    * __Feature extraction__ : 복잡한 차원의 데이터에서 저차원의 데이터로 데이터 변환이 수행되는 feture vector의 처리과정을 말함
    * __Traing set__ : 잠재적인 예측 결과들이 포함되어 있는 데이터셋
    * __Testing set__ : 출력 예측들을 검증하기 위한 데이터 셋

## Different Types of machine learning


* 머신러닝의 학습 방법에는 세가지 타입이 있다.
* __Supervised learning(지도학습)__ : 이미 라벨로 결과를 나타내는 데이터 즉, 타겟 변수가 있을 때 사용하는 학습방법
* __Unsupervised learning(비지도 학습)__ : 타켓 변수가 없을 때 사용하는 학습 방법으로 데이터 내의 패턴을 찾아내는 학습 방법
* __Reinforcement learning(강화 학습)__ : 다양한 환경 요인들과 상호작용하는 프로그램 ex)자동차 자동 운전

### Decision Tree(의사결정나무)
* 입력에 대한 결과를 매핑하여 보여주는 예측 모델 기술임

### Logistic regression(로지스틱 회귀)
* 로지스틱 회귀는 또 다른 지도학습 기술이며, 기본적으로 확률 분류 모델임
* 2개의 예측값을 가지는 변수에 대한 예측에 주로 사용된다.ex) 신용카드 사용내역이 진짜인지 가짜인지 판단
* 로지스틱 회귀는 로지스틱 모형을 사용하며, 로지스틱 함수는 -무한대~+무한대의 범위를 가질 수 있는 유용한 함수이고, 결과 값이 0~1사이에 매핑 되어 0과1의 분류를 확률 모델로 표현할 수 있음
* \\(F(x)=\frac{1}{1+e^-(\beta_0+\beta_1x)}\\)
* 아래의 크림은 로지스틱 ㅎ함수로 부터 도출되는 값의 분포를 보여주며, 분포의 모양은 S커브를 따르는 것을 알 수 있음
