ì´ë¥¼ ì²˜ë¦¬ í•˜ê¸° ìœ„í•´ì„œ NLTKì˜ "Punkt Tokenizer Models"ì„ ì„¤ì¹˜í•´ì•¼ í•œë‹¤. Python3 í™˜ê²½ì„ `python3` ëª…ë ¹ìœ¼ë¡œ ì—´ê³ , `import nltk` ëª…ë ¹ìœ¼ë¡œ nltkë¥¼ import í•œë‹¤. `nltk.download()` ëª…ë ¹ìœ¼ë¡œ NLTKë¥¼ ë‹¤ìš´ë¡œë“œ í•˜ëŠ” ì°½ì„ ì—´ ìˆ˜ ìˆë‹¤. `Models` í˜ì´ì§€ì—ì„œ `punkt - Punkt Tokenizer Models`ì„ ì„ íƒí•˜ë©´ í•´ë‹¹ íŒ¨í‚¤ì§€ë¥¼ ë‹¤ìš´ë¡œë“œ í•  ìˆ˜ ìˆë‹¤.

```python
from nltk.util import ngrams
from nltk import word_tokenize

# word tokenize ë¥¼ ì‚¬ìš©í•˜ì—¬ í…ìŠ¤íŠ¸ë¥¼ ë‹¨ì–´ ë‹¨ìœ„ë¡œ ë‚˜ëˆˆë‹¤
# 3ê°œ ë‹¨ì–´ë¡œ êµ¬ì„±ëœ ngrams ì¦‰, trigramsì„ ì°¾ëŠ”ë‹¤

trigrams = [ngrams(word_tokenize(text), 3) for text in data['text'].values]
trigrams = [item for sublist in trigrams for item in sublist] # flatten list
counter = Counter(trigrams)
for key, value in counter.most_common(10):
    print(key, value)
```

ì‹¤í—˜ ê²°ê³¼ ê°€ì¥ ë§ì´ ë‚˜ì˜¨ trigramsì€ (':', 'https', ':')ë¡œ êµ¬ì„±ëœ ê²ƒê³¼ ('Read', 'more', ':')ë¡œ êµ¬ì„±ëœ ê²ƒì´ì—ˆë‹¤. ì´ì™¸ì— ìì£¼ ë°œìƒí•œ íŒ¨í„´ì€ ('new', 'MacBook', 'Pro'), ('the', 'new', 'MacBook'), ('\$', 'AAPL', '\$') ë“±ì´ ìˆì—ˆë‹¤. ì´ì œ n ì˜ ê°’ì„ 4ë‚˜ 5ë¡œ ë°”ê¾¸ë©´ ì–´ë–¤ íŒ¨í„´ì´ ë‚˜ì˜¤ëŠ”ì§€ë¥¼ ì‚´í´ë³´ê² ë‹¤.

```python
def xgrams(n):
    grams = [ngrams(word_tokenize(text), n) for text in data['text'].values]
    grams = [item for sublist in grams for item in sublist] # flatten list
    counter = Counter(grams)
    for key, value in counter.most_common(10):
        print(key, value)

xgrams(4)
print("\n\n\n")
xgrams(5)
```
