```python
# ì•„ë˜ì—ì„œ '.'ëŠ” í˜„ì¬ ì‘ì—… ë””ë ‰í† ë¦¬ë¥¼ ê°€ë¦¬í‚¨ë‹¤ (ì¦‰, í˜„ì¬ í”„ë¡œê·¸ë¨ì´ ì €ì¥ë˜ì–´ ìˆëŠ”ê³³ì„ ë§í•œë‹¤)
# ì´ë ‡ê²Œ í˜„ì¬ ë””ë ‰í† ë¦¬ë¥¼ ì¤‘ì‹¬ìœ¼ë¡œ í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•¨ìœ¼ë¡œì¨ í”„ë¡œê·¸ë¨ì—ì„œ ì ˆëŒ€ ê²½ë¡œë¥¼ ë‹¤ë£¨ì§€ ì•Šì•„ë„ ë˜ì–´ í¸ë¦¬í•˜ë‹¤
# ì•„ë˜ëŠ” ê³ ì •ëœ ì´ë¦„ì˜ ë””ë ‰í† ë¦¬ë¥¼ ì •ì˜í•˜ëŠ”ë° ì´ì™€ ê°™ì´ ìƒìˆ˜í™”ëœ ë³€ìˆ˜ëŠ” í”„ë¡œê·¸ë¨ì˜ ì²«ë¶€ë¶„ì—ì„œ ì •ì˜í•˜ëŠ” ê²ƒì´ ì¼ë°˜ì ì´ë‹¤

# ì•„ë˜ì—ì„œ os ëŠ” ìš´ì˜ì²´ì œì™€ í†µì‹ í•˜ê¸° ìœ„í•´ ì‚¬ìš©í•˜ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬ì´ë‹¤
# ì´ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ë©´ ìš´ì˜ ì²´ì œì—ê²Œ ì„ì˜ì˜ ëª…ë ¹ë¬¸ì„ ë‚´ë¦´ ìˆ˜ ìˆë‹¤

import os  

path_to_directory = "./data"
path_to_data = os.path.join(path_to_directory, "twitter")

# íŠ¸ìœ„ì„œ ì‚¬ì´íŠ¸ì—ì„œ ì•±ì„ ë“±ë¡í•˜ê³  4ê°€ì§€ í‚¤ ê°’ì„ ë°›ì€ í›„ì— ì´ í‚¤ê°’ë“¤ì„ ì €ì¥í•œ íŒŒì¼ì„ ê°€ë¦¬í‚¤ëŠ” ê²½ë¡œ
# ì„ì˜ë¡œ ì •í•œ í…ìŠ¤íŠ¸ íŒŒì¼ì¸ py-twitter.txtì—ëŠ” ì•„ë˜ì™€ ê°™ì€ ë„¤ê°€ì§€ ì •ë³´ê°€ ìˆœì„œëŒ€ë¡œ ë“¤ì–´ìˆì–´ì•¼ í•œë‹¤
# CONSUMER_KEY
# CONSUMER_SECRET
# ACCESS_TOKEN_KEY
# ACCESS_TOKEN_SECRET

path_to_twitter_keys = "./key/twitter_keys.txt" 

if not os.path.isdir(path_to_directory): # ë°ì´í„° ë””ë ‰í† ë¦¬ê°€ ì—†ìœ¼ë©´ ìƒˆë¡œ ìƒì„±í•œë‹¤
    os.mkdir(path_to_directory)
assert(os.path.exists(path_to_twitter_keys))

# shutilì€ íŒŒì¼ê³¼ ë””ë ‰í„°ë¦¬ë¥¼ ë‹¤ë£¨ëŠ” ìœ í‹¸ë¦¬í‹°ì´ë‹¤
# ê¸°ì¡´ì— ê°™ì€ ì´ë¦„ì˜ ë””ë ‰í† ë¦¬ê°€ ì¡´ì¬í•˜ëŠ”ì§€ í™•ì¸í•˜ê³  ê·¸ ë‚´ìš©ì„ ë¹„ìš°ëŠ” ì‘ì—…ì„ í•œë‹¤
# ì´ë ‡ê²Œ í•˜ëŠ” ëª©ì ì€ ìƒˆë¡œìš´ ë°ì´í„°ì…‹ì„ ì•ˆì „í•˜ê²Œ ì‚¬ìš©í•˜ê¸° ìœ„í•´ì„œì´ë‹¤

import shutil
    
if os.path.isdir(path_to_data):
    shutil.rmtree(path_to_data) # ë””ë ‰í† ë¦¬ì™€ ê·¸ ë‚´ìš©ì„ ëª¨ë‘ ì§€ìš´ë‹¤

os.mkdir(path_to_data)
```

###### 2 ë‹¨ê³„ : ë°ì´í„° ë‹¤ìš´ë¡œë“œ
ì´ì œ ë°ì´í„°ë¥¼ ë‹¤ìš´ë¡œë“œí•œë‹¤. ì§ˆì˜ë¥¼ í•˜ê³  ê²°ê³¼ë¥¼ ë‹¤ìš´ë¡œë“œ í•œë‹¤. ê°™ì€ ë°©ë²•ìœ¼ë¡œ ê³„ì† ë‹¤ìš´ë¡œë“œë¥¼ ì¶”ê°€í•  ìˆ˜ ìˆë‹¤.

** ì£¼ì˜: ë„ˆë¬´ ë§ì€ ì§ˆì˜ë¥¼ ë„ˆë¬´ ìì£¼ í•˜ì§€ ì•Šë„ë¡ ì£¼ì˜í•´ì•¼ í•œë‹¤. ë§¤ 15ë¶„ ë§ˆë‹¤ 150/450 ì§ˆì˜ë¥¼ í•˜ëŠ” ì œí•œì„ ì–´ê¸¸ ê²½ìš° ì§ˆì˜ê°€ ì¼ì‹œì ìœ¼ë¡œ ê¸ˆì§€ë  ìˆ˜ ìˆë‹¤.**

```python
# ë…¸íŠ¸: ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ë©´ ë¬¸ìë¥¼ URL íƒ€ì…ìœ¼ë¡œ ìë™ìœ¼ë¡œ ë³€ê²½í•´ì¤€ë‹¤
# ì˜ˆë¥¼ ë“¤ì–´ "!" ë¥¼ "%21"ë¡œ ë³€ê²½

queries = ['"Apple Inc." OR "#AAPL" OR "$AAPL" OR "Steve Jobs"', 
           '"Tim Cook" OR "Johnny Ive" OR "iPhone" OR "iPad" OR "iOS"', 
           '"Macbook Pro" OR "Macbook Air" OR "Apple Watch"']

def get_twitter_keys():
    # ì§€ì •ëœ í…ìŠ¤íŠ¸ íŒŒì¼ì—ì„œ API í‚¤ ê°’ë“¤ì„ ì½ëŠ”ë‹¤
    # í…ìŠ¤íŠ¸ íŒŒì¼ì—ì„œ ì½ì§€ ì•ŠëŠ” ê²½ìš° ì´ë¶€ë¶„ì˜ ì½”ë“œëŠ” ë³€ê²½ë˜ì–´ì•¼ í•œë‹¤
    with open(path_to_twitter_keys, "r") as f:
        consumer_key = f.readline().strip() # stripì€ ë¹ˆ ê³µê°„, ìŠ¤í˜ì´ìŠ¤ë‚˜ íƒ­ì„ ì œê±°í•œë‹¤
        consumer_secret = f.readline().strip() 
        access_token = f.readline().strip()
        access_token_secret = f.readline().strip()
        return consumer_key, consumer_secret, access_token, access_token_secret

# Tywthonì— ëŒ€í•œ ìì„¸í•œ ë‚´ìš© - https://twython.readthedocs.io/en/latest/index.html
# pythonì´ ì„¤ì¹˜ëœ ê³³ì—ì„œ 'pip install twython'ìœ¼ë¡œ ì„¤ì¹˜í•  ìˆ˜ ìˆë‹¤

from twython import Twython
    
consumer_key, consumer_secret, access_token, access_token_secret = get_twitter_keys()
twitter = Twython(consumer_key, consumer_secret, access_token, access_token_secret)

# ì§ˆì˜ë¥¼ í†µí•´ì„œ ì–»ëŠ” ëª¨ë“  íŠ¸ìœ— ë°ì´í„°ë¥¼ í•˜ë‚˜ì˜ ë¦¬ìŠ¤íŠ¸ì— ë‹´ëŠ”ë‹¤
all_tweets = []
for query in queries:
    response = twitter.search(q=query, count=100, lang='en')
    metadata = response['search_metadata'] 
    
    # dict_keys(['max_id_str', 'completed_in', 'next_results', 'max_id', 'refresh_url', 
    #'query', 'since_id_str', 'count', 'since_id'])
    
    tweets = response['statuses'] 
    
    """ list of dicts (tweets): dict_keys(['lang', 'retweeted', 'retweet_count', 'truncated',
    'in_reply_to_status_id', 'place', 'in_reply_to_screen_name', 'id_str', 'geo', 'text', 
    'favorite_count', 'in_reply_to_user_id', 'created_at', 'in_reply_to_status_id_str', 
    'coordinates', 'source', 'entities', 'user', 'in_reply_to_user_id_str', 'metadata', 
    'favorited', 'contributors', 'is_quote_status', 'id'])
    tweets contains count=100 tweets. for the next page, you need another twitter query with 
    the query string stored in metadata['next_results']. 
    if you do multiple pages, beware the limit on queries per time period (currently 150/15m 
    for individual users, 450/15m for applications)
    """
    all_tweets += tweets
        
    # íŠ¸ìœ„í„°ëŠ” ì§ˆì˜ì— í•´ë‹¹í•˜ëŠ” ê²°ê³¼ë¥¼ ì†Œìˆ˜ë§Œ ë¦¬í„´í•œë‹¤ 
    while "next_results" in metadata.keys():
        response = twitter.search(q=metadata['next_results'], count=100, lang='en')
        metadata = response['search_metadata'] 
        # dict_keys(['max_id_str', 'completed_in', 'next_results', 'max_id', 'refresh_url', 
        #'query', 'since_id_str', 'count', 'since_id'])
        tweets = response['statuses']
        
        all_tweets += tweets
   
assert(len(all_tweets) >= 1) # ìµœì†Œí•œ 1ê°œ ì´ìƒì˜ ê²°ê³¼ë¥¼ ì–»ì—ˆëŠ”ì§€ í™•ì¸

# ê²°ê³¼ë¡œ ì–»ëŠ” íŠ¸ìœ—ì˜ ë‚´ìš©ì„ ë³¼ ìˆ˜ ìˆëŠ”ë°, ì˜ì–´ê°€ ì•„ë‹ ìˆ˜ë„ ìˆë‹¤
# ì–¸ì–´ë¥¼ êµ¬ë¶„í•˜ëŠ” ê¸°ëŠ¥ì´ ì™„ì „í•˜ì§€ëŠ” ì•Šë‹¤

for tweet in all_tweets[:10]:
    print(tweet['text'])
    
# ê²°ê³¼ íŠ¸ìœ—ì„ íŒŒì¼ì— ì €ì¥í•˜ì—¬ ë¶„ì„ì„ í¸ë¦¬í•˜ê²Œ ìˆ˜í–‰í•œë‹¤
# Pandasë¥¼ ì‚¬ìš©í•˜ì—¬ íŠ¸ìœ—ì„ Pandas Dataframeìœ¼ë¡œ ë°”ê¾¸ê³  ì´ë¥¼ csv íŒŒì¼ì— ê¸°ë¡í•œë‹¤

name = "tweets.csv"
path = os.path.join(path_to_data, name)

import pandas as pd
data = pd.DataFrame.from_dict(all_tweets)
data.to_csv(path)
```

ì•ì—ì„œ ë°°ìš´ ë‚´ìš©ì˜ ì½”ë“œë¥¼ ì¬êµ¬ì„±í•˜ì—¬ ê°„ê²°í•œ ì½”ë“œë¡œ ë§Œë“¤ì—ˆê³  ë°ì´í„°ë¥¼ ì–»ëŠ” ë©”ì¸ í•¨ìˆ˜ëŠ” ì•„ë˜ì™€ ê°™ì´ í˜¸ì¶œí•  ìˆ˜ ìˆë‹¤.

```python
import os
import requests
import pandas as pd
from twython import Twython

path_to_directory = "./data"
path_to_data = os.path.join(path_to_directory, "twitter")

path_to_twitter_keys = "./key/twitter_keys.txt"
assert(os.path.exists(path_to_twitter_keys))

name = "twitter.csv"
path = os.path.join(path_to_data, name)

def check_destination():
    if not os.path.isdir(path_to_directory):
        os.mkdir(path_to_directory)
    
    if not os.path.isdir(path_to_data):
        os.mkdir(path_to_data)

def get_twitter_keys():
    # ì§€ì •ëœ í…ìŠ¤íŠ¸ íŒŒì¼ì—ì„œ API í‚¤ ê°’ë“¤ì„ ì½ëŠ”ë‹¤
    # í…ìŠ¤íŠ¸ íŒŒì¼ì—ì„œ ì½ì§€ ì•ŠëŠ” ê²½ìš° ì´ë¶€ë¶„ì˜ ì½”ë“œëŠ” ë³€ê²½ë˜ì–´ì•¼ í•œë‹¤

    with open(path_to_twitter_keys, "r") as f:
        consumer_key = f.readline().strip() # stripì€ ë¹ˆ ê³µê°„, ìŠ¤í˜ì´ìŠ¤ë‚˜ íƒ­ì„ ì œê±°í•œë‹¤
        consumer_secret = f.readline().strip() 
        access_token = f.readline().strip()
        access_token_secret = f.readline().strip()
        return consumer_key, consumer_secret, access_token, access_token_secret
        
def download_data(): 
    queries = ['"Apple Inc." OR "#AAPL" OR "$AAPL" OR "Steve Jobs"', 
           '"Tim Cook" OR "Johnny Ive" OR "iPhone" OR "iPad" OR "iOS"', 
           '"Macbook Pro" OR "Macbook Air" OR "Apple Watch"']
    consumer_key, consumer_secret, access_token, access_token_secret = get_twitter_keys()
    twitter = Twython(consumer_key, consumer_secret, access_token, access_token_secret)

    all_tweets = []
    for query in queries:
        response = twitter.search(q=query, count=100, lang='en')
        metadata = response['search_metadata']
        tweets = response['statuses']
        all_tweets += tweets

        while "next_results" in metadata.keys():
            response = twitter.search(q=metadata['next_results'], count=100, lang='en')
            metadata = response['search_metadata']
            tweets = response['statuses']

            all_tweets += tweets
    
    data = pd.DataFrame.from_dict(all_tweets)
    data.to_csv(path)
    
def verify_data(data):
    columns = data.columns.values.tolist()
    required_columns = ['lang', 'retweeted', 'retweet_count', 'truncated', 'in_reply_to_status_id', 'place', 'in_reply_to_screen_name', 'id_str', 'geo', 'text', 'favorite_count', 'in_reply_to_user_id', 'created_at', 'in_reply_to_status_id_str', 'coordinates', 'source', 'entities', 'user', 'in_reply_to_user_id_str', 'metadata', 'favorited', 'contributors', 'is_quote_status', 'id']
    for column in required_columns:
        assert(column in columns)

    languages = set(data['lang'].values.tolist())
    assert(languages == set(['en'])) # ì˜ì–´ë¡œ ëœ íŠ¸ìœ—ë§Œ ìˆëŠ”ì§€ í™•ì¸
            
def load_tweets(overwrite=False):
    if overwrite or not os.path.exists(path):
        check_destination()
        download_data()
    
    data = pd.read_csv(path)
    verify_data(data)
    return data
```
