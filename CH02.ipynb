## Binomial Distribution(이항 분포)

Plotting the binomial distribution

```python
fig, ax = plt.subplots(1, 1)
x = [0, 1, 2, 3, 4, 5, 6]
n, p = 6, 0.5
rv = stats.binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
          label='Probablity')
ax.legend(loc='best', frameon=False)
plt.xlabel('No. of instances')
plt.ylabel('Probability')
plt.show()
```

## Z-Score

Generating classroom scores and viewing the distribution

```python
classscore = np.random.normal(50, 10, 60).round()

plt.hist(classscore, 20, normed=True)
plt.xlabel('No. of students')
plt.show()
```

Finding the z-score of each of the marks

```python
stats.zscore(classscore)
```

Finding the probability of values higher than 1.334 z-score

```python
prob = 1 - stats.norm.cdf(1.334)
prob
```

Finding the Z-score of top 20% marks

```python
stats.norm.ppf(0.80)
```

Finding the value of the Z-score of 0.84

```python
(0.84 * classscore.std()) + classscore.mean()
```

## P-Value

Is it common to score 75 marks in mathematics. Let's find the z-score of it.

```python
zscore = ( 68 - classscore.mean() ) / classscore.std()
zscore
```

Let's convert the probability to z-score

```python
prob = 1 - stats.norm.cdf(zscore)
prob
```

## Confidence Interval

Let’s try to understand this concept by using an example. Lets take the height of every man in Kenya and determine with 95% confidence interval the average of height of Kenyan men at a national level.
Let’s take 50 men and their height in centimeters

```python
height_data = np.array([186.0, 180.0, 195.0, 189.0, 191.0, 177.0, 161.0, 177.0, 192.0, 182.0, 185.0, 192.0,
                        173.0, 172.0, 191.0, 184.0, 193.0, 182.0, 190.0, 185.0, 181.0, 188.0, 179.0, 188.0,
                        170.0, 179.0, 180.0, 189.0, 188.0, 185.0, 170.0, 197.0, 187.0, 182.0, 173.0, 179.0,
                        184.0, 177.0, 190.0, 174.0, 203.0, 206.0, 173.0, 169.0, 178.0, 201.0, 198.0, 166.0,
                        171.0, 180.0])

plt.hist(height_data, 30, normed=True)
plt.xlabel('Height')
plt.show()
```

The mean of the distribution is

```python
height_data.mean()
```

So the average height of a man from the sample is 183.4 cm

To determine the confidence interval, we’ll now define the standard error of the mean. 

The standard error of the mean is the deviation of the sample mean can have from the population mean. It is defined as the standard deviation by the root of the number of elements.

```python
stats.sem(height_data)
```

So there is a standard error of the mean of 1.38 cm. So the lower and upper limit of the confidence interval will be

<b>Lower Limit</b>: 183.24 + (1.96 * 1.38) = 185.94

<b>Upper Limit</b>: 183.24 - (1.96*1.38) =  180.53

1.96 standard deviation covers 95% of area in the normal distribution

So we can confidently say that the population mean lies in between 180.53 cm and 185.94 cm of height.

Lets assume, we take a sample of 50 people, record their height and then repeat this process 30 times. We can then plot the averages of each sample and observe the distribution.

```python
average_height = []
for i in xrange(30):
    sample50 = np.random.normal(183, 10, 50).round()
    average_height.append(sample50.mean())

plt.hist(average_height, 10, normed=True)
plt.xlabel('Height')
plt.show()
```

You can observe that the mean ranges from 180 to 187 cm when we simulated the average height of 50 sample men which was taken 30 times. 

Let’s, see what happens when we 1000 sample men and repeat the process 30 times

```python
average_height = []
for i in xrange(30):
    sample1000 = np.random.normal(183, 10, 1000).round()
    average_height.append(sample1000.mean())

plt.hist(average_height, 10, normed=True)
plt.xlabel('Height')
plt.show()
```

As you can see, the height varies from 182 and to 185. What does this mean?
It means that as sample size increases, the standard error of the mean decreases which also means that the confidence interval becomes more narrow and we can tell with certainty that at what interval the population mean would lie which is pretty cool

## Correlation

. Let’s take the mileage and horsepower of various cars and see if there is a relation between it. This can be achieved using pearsonr function in the scipy package

```python
mpg = [21.0, 21.0, 22.8, 21.4, 18.7, 18.1, 14.3, 24.4, 22.8, 19.2, 17.8, 16.4, 17.3, 15.2, 10.4, 10.4, 14.7, 32.4, 30.4,
       33.9, 21.5, 15.5, 15.2, 13.3, 19.2, 27.3, 26.0, 30.4, 15.8, 19.7, 15.0, 21.4]
hp = [110, 110, 93, 110, 175, 105, 245, 62, 95, 123, 123, 180, 180, 180, 205, 215, 230, 66, 52, 65, 97, 150, 150, 245,
      175, 66, 91, 113, 264, 175, 335, 109]

stats.pearsonr(mpg, hp)

plt.scatter(mpg, hp)
plt.xlabel('mpg')
plt.ylabel('hp')
plt.show()
```

```python
stats.pearsonr(mpg,hp)
```

The 1st value of the output gives the correlation between the horsepower and the mileage and 2nd value gives the p-value.

So, the 1st value tells us that it is highly negatively correlated and the p-value tells us that there is significant correlation between them.

```python
stats.spearmanr(mpg,hp)
```

We can see that the spearman correlation is -0.89 and the p-value is significant.

Let’s do an experiment in which we introduce few outlier values in the data and see how Pearson and Spearman correlation gets affected.

```python
mpg = [21.0, 21.0, 22.8, 21.4, 18.7, 18.1, 14.3, 24.4, 22.8, 19.2, 17.8, 16.4, 17.3, 15.2, 10.4, 10.4, 14.7, 32.4, 30.4,
       33.9, 21.5, 15.5, 15.2, 13.3, 19.2, 27.3, 26.0, 30.4, 15.8, 19.7, 15.0, 21.4, 120, 3]
hp = [110, 110, 93, 110, 175, 105, 245, 62, 95, 123, 123, 180, 180, 180, 205, 215, 230, 66, 52, 65, 97, 150, 150, 245,
      175, 66, 91, 113, 264, 175, 335, 109, 30, 600]

plt.scatter(mpg, hp)
plt.xlabel('mpg')
plt.ylabel('hp')
plt.show()
```

```python
stats.pearsonr(mpg, hp)
```

```python
stats.spearmanr(mpg, hp)
```

We can clearly see that the pearson correlation has got drastically affected due to the outliers which is from a correlation of 0.89 to a  0.47.  

The spearman correlation did not got affected much as it is based on the order rather than the value.

## T-Test

T-test is used for comparing the population mean against a sample or comparing the population mean of two distributions with sample size less than 30 and when you don’t know the population’s standard deviation.


Let’s do a T-test on two classes that given a mathematics and have 10 students in each class

```python
class1_score = np.array([45.0, 40.0, 49.0, 52.0, 54.0, 64.0, 36.0, 41.0, 42.0, 34.0])
class2_score = np.array([75.0, 85.0, 53.0, 70.0, 72.0, 93.0, 61.0, 65.0, 65.0, 72.0])

stats.ttest_ind(class1_score, class2_score)
```

The 1st value in the output is the calculated t-statistics whereas the 2nd value is the p-value and p-value shows that the two distributions are not identical

## Chi-Square

Chi-Square for Goodness of fit

Chi-Square test can be used to test whether the observed data differs significantly from the expected data.

Let’s take the example of a dice.

The dice is rolled 36 times and the probability that each face should turn up is ⅙. So the expected distribution is as follows

```python
expected = np.array([6, 6, 6, 6, 6, 6])
observed = np.array([7, 5, 3, 9, 6, 6])

stats.chisquare(observed, expected)
```

The 1st value is the chi-square value and the 2nd value is the p-value which is very high. This means that the null hypothesis is valid and observed is similar to the expected

Chi-Square Test of Independence

The Chi-Square Test of independence is a statistical test to determine if two categorical variables are independent of each other or not

Let’s take the following example to see if there is a preference between gender and genre of a book

```python
men_women = np.array([[100, 120, 60],[350, 200, 90]])
stats.chi2_contingency(men_women)
```

The 1st value is the chi-square value. 

The 2nd value is the p-value which is very small which means that there is an association between the gender and genre of the book they read.

The 3rd value is the degrees of freedom

The 4th value which is an array is the expected frequencies

ANOVA is a test that cannot tell you which group is significantly different from each other.
